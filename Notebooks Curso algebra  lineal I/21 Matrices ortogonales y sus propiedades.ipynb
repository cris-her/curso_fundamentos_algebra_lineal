```
import numpy as np
```

```
matriz = np.array([[1,0,0],
                   [0,1,0],
                   [0,0,1],
                   ])

print(matriz)
```

```
# Comprobacion v1 ortogonal v2
print(matriz[:,0].dot(matriz[:,1]))
```

```
# Comprobacion v1 ortogonal v3
print(matriz[:,0].dot(matriz[:,2]))
```

```
# Comprobacion v2 ortogonal v3
print(matriz[:,1].dot(matriz[:,2]))
```

```
# Comprobar las filas son ortonormales
print(np.linalg.norm(matriz[:,0]))
print(np.linalg.norm(matriz[:,1]))
print(np.linalg.norm(matriz[:,2]))
```

```
# Comprobar las columnas son ortogonales
print(matriz[0,:].dot(matriz[1,:]))
print(matriz[0,:].dot(matriz[2,:]))
print(matriz[1,:].dot(matriz[2,:]))
```

```
# Comprobar las columnas son ortonormales
print(np.linalg.norm(matriz[0,:]))
print(np.linalg.norm(matriz[1,:]))
print(np.linalg.norm(matriz[2,:]))
```

Las matrices son todas ortogonales con vectores ortonormales, no existe la definicion de matriz ortonormal

```
#utilizando una matriz de rotacion. 

A = np.array([[np.cos(100), -np.sin(100)],
             [np.sin(100), np.cos(100)]])

print(A)
```

```
print(np.linalg.norm(A[0,:]))
print(np.linalg.norm(A[1,:]))
print(np.linalg.norm(A[:,0]))
print(np.linalg.norm(A[:,1]))
```

```
print(A[0,:].dot(A[1,:]))
print(A[:,0].dot(A[:,1]))
```

```
A_t = A.T
print(A_t.dot(A))
print(A.dot(A_t))
```

Esta es la matriz identidad, la maquina toma el e-18 como si fuera un numero cercano a cero

```
print(1/A_t.dot(A))
```

AL hacer esto el numero cercano a cero se transforma en un numero muy grande, intentando calcular  una inversa. Debemos tener cuidado para no expandir nuestros errores.
