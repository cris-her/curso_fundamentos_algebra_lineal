```
import numpy as np
```

```
vector = np.array([1,2,3,4,5])

matriz1 = np.diag(vector)

print(matriz1)
```

```
print(matriz1[0:4,0:3])
```

```
print(matriz1[0:3,0:4])
```

```
A = np.diag([2,3,4,5])
print(A)
```

```
v1 = np.array([[1,1,1,1]])

print(v1)

print(A.dot(v1.T))
```

### Matriz inversa

```
# Obtenemos la inversa dividiendo entre los elementos de la diagonal
# de la matriz A
A_inv = np.diag([1/2, 1/3, 1/4, 1/5])
print(A_inv)
```

```
identidad = A.dot(A_inv)
print(identidad)
```

```
A_inv_calc = np.linalg.inv(A)
print(A_inv_calc)
```
