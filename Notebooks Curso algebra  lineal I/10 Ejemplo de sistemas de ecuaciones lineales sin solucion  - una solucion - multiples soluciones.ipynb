```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

## Sistema sin solucion (sobredeterminado)


```
x = np.arange(-6, 6)

y_1 = 3*x + 5
y_2 = -x + 3
y_3 = 2*x + 1
```

```
plt.figure()
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-14,14)
plt.ylim(-14,14)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```

Este es un sistema sobredeterminado: tenemos mas ecuaciones que variables, para que sea cierto el sistema las tres rectas deben tocarse en el mismo punto.


## Sistema una solucion (determinado)

```
x = np.arange(-6, 6)

y_2 = -x + 3
y_3 = 2*x + 1
```

```
plt.figure()
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.xlim(-14,14)
plt.ylim(-14,14)

plt.axvline(x=0, color='grey')
plt.axhline(y=0, color='grey')
plt.show()
```
