```
import numpy as np
```

```
A = np.array([[2,3], [5,7], [11,13]])
B = np.array([[1,3], [2,1]])
C = np.array([[3,1], [4,2]])
```

```
# Asociativa
ABC = A.dot(B.dot(C))
print(ABC)
```

```
AB_C = A.dot(B).dot(C)
print(AB_C)
```

```
# Distributiva
D = A.dot(B+C)
E = (A.dot(B)) + (A.dot(C))
```

```
print(D)
print(E)
```

```
print(D==E)
```

```
F = B.dot(C)
```

```
G = C.dot(B)
```

```
print(F==G)
```

```
# No fue conmutativa
print(F)
```

```
print(G)
```
